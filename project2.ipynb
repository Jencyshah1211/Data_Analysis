```python
# Top 5 countries distribution using barplot
li = list(file['country'].value_counts()[0:5].keys())
plt.figure(figsize=(8,5))
plt.bar(li, list(file['country'].value_counts()[0:5]),color = "y")
plt.show()
```

```python
# Top 5 countries distribution using pie-chart
plt.figure(figsize=(7,7))
plt.pie(list(file['country'].value_counts()[0:5]),labels=list(file['country'].value_counts()[0:5].keys()),autopct = '%0.1f%%')
plt.show()
```

```python
# Distribution of top 5 ratings via bar-plot
li2 = list(file['rating'].value_counts()[0:5].keys())
plt.figure(figsize=(8,5))
plt.bar(li2, list(file['rating'].value_counts()[0:5]),color = "r")
plt.show()
```

```python
# Distribution of top 5 ratings via pie-chart
plt.figure(figsize=(7,7))
plt.pie(list(file['rating'].value_counts()[0:5]),labels=list(file['rating'].value_counts()[0:5].keys()),autopct = '%0.1f%%')
plt.show()
```

```python
# Top 5 release_year distribution using barplot
li3 = list(file['release_year'].value_counts()[0:5].keys())
plt.figure(figsize=(8,5))
plt.bar(li3, list(file['release_year'].value_counts()[0:5]),color = "g")
plt.show()
```

```python
# Distribution of top 5 release year via pie-chart
plt.figure(figsize=(7,7))
plt.pie(list(file['release_year'].value_counts()[0:5]),labels=list(file['release_year'].value_counts()[0:5].keys()),autopct = '%0.1f%%')
plt.show()
```

```python
# Type distribution using barplot
li4 = list(file['type'].value_counts().keys())
plt.figure(figsize=(8,5))
plt.bar(li4, list(file['type'].value_counts()),color = "brown")
plt.show()
```
